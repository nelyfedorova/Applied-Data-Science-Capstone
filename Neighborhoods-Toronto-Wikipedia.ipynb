# Reading Toronto data from Wikipedia 

To create a dataframe from Toronto data:

- The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood
- Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
- More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the neighborhoods separated with a comma as shown in row 11 in the above table.
- If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough. So for the 9th cell in the table on the Wikipedia page, the value of the Borough and the Neighborhood columns will be Queen's Park.

```python
import pandas as pd
```

Read Wikipedia page and extract the table

```python
# Link To Wikipedia page
path='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

# Read File
df_html = pd.read_html(path)

# Extract the table 
df = df_html[0]

# Rename Columns
df.rename(columns={0:'Postcode', 1: 'Borough', 2: 'Neighborhood'}, inplace=True)

# Eliminate the first row
df = df.drop([0])

# Ignore cells with a borough that is Not assigned
df = df[df.Borough != 'Not assigned']
print(df.shape)
```

Combine rows with same Postcode into comma-separated lists

```python
df.set_index(['Postcode','Borough'],inplace=True)
df1 = df.groupby(level=['Postcode','Borough'], sort=False).agg( ','.join)
print(df1.shape)
df1
```

If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough

```python
df1 = df1.reset_index()
#df1[df1['Neighborhood'] == 'Not assigned']
df1.loc[4, 'Neighborhood'] = df1.loc[4, 'Borough'] # Queen's Park
df1.head()
```

Save to xls file

```python
df1.to_excel('wiki_toronto_data.xls')
```
